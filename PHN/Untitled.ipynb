## 제주 사회 초년생의 금융 소비 행동 양상 

```python
# 모듈 import
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as sm
import numpy as np
```

```python
df = pd.read_csv('jeju_financial_life_data .csv')
```

```python
#OLS모델로 변수간 상관관례 분석
age_behavior = sm.ols(formula = 'age~avg_income+avg_spend+avg_debt', data = df).fit()
```

```python
age_behavior.summary()
```
